```python
import pickle
import codecs
import json
from collections import defaultdict
```

```python
tweets_pickle = "/home/oscar/trend-tracker/tools/tweets.pickle"
hashtags_db_path = "/home/oscar/trend-tracker/tools/hashtags_db.pickle"
h_db_js = "/home/oscar/trend-tracker/js/hashtags.js"

def searchHashtags():
    with open(tweets_pickle, 'r') as f:
        tweets = pickle.load(f)
    print "Number of new tweets", len(tweets)
    hashtags = defaultdict(int)
    for tweet in tweets:
        tweet_hashtag = tweet['entities']['hashtags']
        if len(tweet_hashtag) > 0:
            for hashtag in tweet_hashtag:
                hashtags[hashtag['text']] += 1
    return hashtags

def write_hashtags_db(h_db):
    with open(hashtags_db_path, 'w') as f:
        pickle.dump(hashtags_db, f)
        
def save_to_hashtags_db(t_h):
    #with open(hashtags_db_path, 'r') as f:
    #    h_db = pickle.load(f)
    h_db = defaultdict(int)
    for hashtag in t_h:
        h_db[hashtag] += 1 
    return h_db

def hashtags_db_to_js(h_db):
    hashtags_weight = []
    count = 0
    for hashtag in h_db:
        count += h_db[hashtag]
        
    for hashtag in h_db:
        # weight = float(h_db[hashtag] / float(count))
        weight = float(h_db[hashtag])
        print weight
        h = {'text': hashtag, 'weight': weight}
        hashtags_weight.append(h)
        
    tweets_json = json.dumps(hashtags_weight)
    print tweets_json
    with codecs.open(h_db_js, 'w', 'utf-8') as f:
        f.write('var hashtags = ' +"'"+tweets_json+"'" + ';\n var totalCount = ' + str(count))
    print 'Saved', len(h_db), 'hashtags into', h_db_js, 'with a count of', str(count)
```
